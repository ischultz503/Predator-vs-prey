```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pandas as pd
from __future__ import print_function
from ipywidgets import interact
import ipywidgets as widgets
%matplotlib inline
```

```python
a = np.log(2)
b = np.log(2)/5
c = np.log(2)
d = np.log(2)/20
ts = np.linspace(1,200)
```

```python
def dP_dt (P , t):
    return [P[0]*(a-b*P[1]), -P[1]*(c-d*P[0])]
```

```python
def ode_solver(eqn, m0, ts):
    P0 = [m0 , s0]
    Ps=odeint(dP_dt , P0 , ts)
    return Ps
```

```python
def time_series_plot():
    plt.plot(ts, population['prey'], "-", label = "Tiny Turtles")
    plt.plot(ts, population['predators'], "g", label = "Monster Meerkats")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.legend();
```

```python
def versus_plot():
    plt.plot(population['prey'] , population['predators'] , "c")
    plt.xlabel("Mighty Meerkats")
    plt.ylabel("Terrible Turtles")
    plt.title("Meerkats vs. Turtles");
```

```python
def pop_solve(m0):
    prey = ode_solver(dP_dt , m0 , ts)[:,0]
    predators = ode_solver(dP_dt , m0 , ts)[:,1]
    population = pd.DataFrame({'prey':prey, 'predators':predators} , dtype = int)
    return population
```

```python
def fun_plot(m0):
    prey = ode_solver(dP_dt , m0 , ts)[:,0]
    predators = ode_solver(dP_dt , m0 , ts)[:,1]
    plt.plot(ts, prey, "-", label = "Tiny Turtles")
    plt.plot(ts, predators, "g", label = "Monster Meerkats")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.legend()
```
