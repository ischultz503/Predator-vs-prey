```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pandas as pd
%matplotlib inline
```

```python
#a = np.log(2); b = np.log(2)/5; c = np.log(2); d = np.log(2)/20;
```

```python
a,b,c,d = 1,1,1,1; m0=3;s0=5;
```

```python
def dP_dt (P, t):
    return [P[0]*(a-b*P[1]), -P[1]*(c-d*P[0])]

ts = np.linspace(1,200)
P0=[m0, s0]
Ps = odeint(dP_dt, P0, ts)
prey = Ps[:,0]
predators = Ps[:,1]
```

```python
plt.plot(ts, prey, "-", label = "Tiny Turtles")
plt.plot(ts, predators, "g", label = "Monster Meerkats")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend();
```

```python
plt.plot(prey , predators , "c")
plt.xlabel("Mighty Meerkats")
plt.ylabel("Terrible Turtles")
plt.title("Meerkats vs. Turtles");
```
